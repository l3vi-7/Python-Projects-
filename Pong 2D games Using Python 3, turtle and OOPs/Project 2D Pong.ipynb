```python
#Pong game using Python3, turtle module and OOPs concepts.
# by Shravan Shetty

import turtle
import winsound

game_window = turtle.Screen()
game_window.title("Project Pong")
game_window.bgcolor("black")
game_window.setup(width = 600, height = 400)
game_window.tracer(0)


#Player 1
player_1 = turtle.Turtle()
player_1.speed(0)
player_1.shape("square")
player_1.color("white")
player_1.shapesize(stretch_wid = 5, stretch_len = 1)
player_1.penup()
player_1.goto(-250, 0)

#Player 2
player_2 = turtle.Turtle()
player_2.speed(0)
player_2.shape("square")
player_2.color("white")
player_2.shapesize(stretch_wid = 5, stretch_len = 1)
player_2.penup()
player_2.goto(250, 0)

#Pong
ball = turtle.Turtle()
ball.speed(0)
ball.shape("square")
ball.color("white")
ball.penup()
ball.goto(0, 0)
ball.dx = 0.1
ball.dy = -0.1

#Movement class
class Player_movement:
    def __init__(self, player):
        self.player = player
        
    def up(self):
        y = (self.player).ycor()
        if y <= 200:
            y+= 20
        (self.player).sety(y)
    
    def down(self):
        y = (self.player).ycor()
        if y >= -200:
            y -= 20
        (self.player).sety(y)
        
#Player Movement Object 
        
A = Player_movement(player_1)
B = Player_movement(player_2)

#Movevement

game_window.listen()
game_window.onkeypress(A.up, "w")
game_window.onkeypress(A.down, "s")
game_window.onkeypress(B.up, "8")
game_window.onkeypress(B.down, "2")

#Scores
player1_scores = 0
player2_scores = 0

#Score Board
score_board = turtle.Turtle()
score_board.speed(0)
score_board.color("Blue")
score_board.penup()
score_board.hideturtle()
score_board.goto (0, 180)
score_board.write("Player 1: 0", align="center", font=("Calibri", 15, "normal"))
score_board.color("Red")
score_board.goto (0, 160)
score_board.write("Player 2: 0", align="center", font=("Calibri", 15, "normal"))


# Main game
while True:
    game_window.update()
    # Ball Movement Initialization
    ball.setx(ball.xcor() + ball.dx)
    ball.sety(ball.ycor() + ball.dy)
    
    # Boundary Conditions
    
    if ball.ycor() > 190:
        ball.sety(190)
        ball.dy *= -1
        winsound.PlaySound("pong.mp3", winsound.SND_ASYNC)
    
    if ball.ycor() < -190 :
        ball.sety(-190)
        ball.dy *= -1
        winsound.PlaySound("pong.mp3", winsound.SND_ASYNC)
        
    if ball.xcor() > 290:
        ball.goto(0, 0)
        ball.dx *= -1
        player1_scores += 1
        score_board.clear()
        score_board.goto (0, 180)
        score_board.color("Blue")
        score_board.write("Player 1: {}".format(player1_scores), align="center", font=("Calibri", 15, "normal"))
        score_board.color("Red")
        score_board.goto (0, 160)
        score_board.write("Player 2: {}".format(player2_scores), align="center", font=("Calibri", 15, "normal"))

    if ball.xcor() < -290:
        ball.goto(0, 0)
        #ball.setx(-270)
        ball.dx *= -1
        player2_scores +=1
        score_board.clear()
        score_board.goto (0, 180)
        score_board.color("Blue")
        score_board.write("Player 1: {}".format(player1_scores), align="center", font=("Calibri", 15, "normal"))
        score_board.color("Red")
        score_board.goto (0, 160)
        score_board.write("Player 2: {}".format(player2_scores), align="center", font=("Calibri", 15, "normal"))
    #GamePlay
    
    if (ball.xcor() > 240 and ball.xcor() < 250) and (ball.ycor() < player_2.ycor() + 50 and ball.ycor() > player_2.ycor() - 50):
        ball.setx(240)
        ball.dx *= -1
    
    if (ball.xcor() < -240 and ball.xcor() > -250) and (ball.ycor() < player_1.ycor() + 50 and ball.ycor() > player_1.ycor() - 50):
        ball.setx(-240)
        ball.dx *= -1
        
    #Winning Condition
    
    if player1_scores == 4 :
        score_board.clear()
        score_board.goto (0, 150)
        score_board.color("Blue")
        score_board.write("Player 1: {}".format(player1_scores), align="center", font=("Calibri", 15, "normal"))
        winsound.PlaySound("pong.mp3", winsound.SND_ASYNC)
        winsound.PlaySound("pong.mp3", winsound.SND_ASYNC)
        score_board.goto (0,0)
        score_board.color("White")
        score_board.write("Player 1 Wins!!!", align="center", font=("Calibri", 22, "bold"))
        winsound.PlaySound("pong.mp3", winsound.SND_ASYNC)
        winsound.PlaySound("pong.mp3", winsound.SND_ASYNC)
        ball.goto(0,-150)
        ball.dx = 0
        ball.dy = 0
    
    if player2_scores == 4 :
        score_board.clear()
        score_board.goto (0, 150)
        score_board.color("Red")
        score_board.write("Player 2: {}".format(player2_scores), align="center", font=("Calibri", 22, "normal"))
        winsound.PlaySound("pong.mp3", winsound.SND_ASYNC)
        winsound.PlaySound("pong.mp3", winsound.SND_ASYNC)
        score_board.goto (0,0)
        score_board.color("White")
        score_board.write("Player 2 Wins!!!", align="center", font=("Calibri", 30, "bold"))
        ball.goto(0,-150)
        ball.dx = 0
        ball.dy = 0
        
        
    
        

    
```
